### 2. BIDMC - Apply all the summary info (checked previously)

```python
# Used non_diagnostic mastersheet as base
summary = non_diagnostic_mastersheet[['sub_id', 'session', 'diagnostic']].copy()

# Add info about EDF and ECG (all have ECG)
summary['has_edf'] = 1
summary['has_ecg'] = summary['diagnostic'].astype(int)

# Add the 1 PSG without EDF
no_edf_row = pd.DataFrame({
    "sub_id": ["I0002150013905"],
    "session": [1],
    "has_edf": 0,
    "diagnostic": 0,
    "has_ecg":0,
})
summary = pd.concat([summary, no_edf_row], ignore_index=True)

# long_recording if more than 3h of recording
long_sessions  = non_diagnostic_mastersheet.loc[
    non_diagnostic_mastersheet['duration_sec'] >= 3*3600,
    ['sub_id', 'session']
].set_index(['sub_id', 'session']).index
mask_duration = summary.set_index(['sub_id', 'session']).index.isin(long_sessions)
summary['long_recording'] = ((
    summary['has_ecg'] == 1) & mask_duration).astype(int)
              
# Has sleep stage if annot path and long_recording
has_annot = diagnostic_mastersheet.loc[
    ~diagnostic_mastersheet['sleep_stage_path'].isna(),
    ['sub_id', 'session']
].set_index(['sub_id', 'session']).index
mask_has_annot = summary.set_index(['sub_id', 'session']).index.isin(has_annot)
summary['has_sleep_stages'] = ((summary['long_recording'] == 1) & mask_has_annot).astype(int)


# Not the one that do not have valid stages
summary['has_valid_sleep_stages'] = 1
mask_no_valid_stages = summary.set_index(['sub_id', 'session']).index.isin(no_valid_stages)
mask_has_sleep_0 = summary['has_sleep_stages'] == 0
summary.loc[mask_no_valid_stages | mask_has_sleep_0, 'has_valid_sleep_stages'] = 0

# Add also the too_early_start
summary['has_correct_start'] = (summary['has_valid_sleep_stages'] == 1).astype(int)

# Add info for too long sleep stages
summary['has_correct_length_stages'] = 1
mask_too_long_stages = summary.set_index(['sub_id', 'session']).index.isin(too_long_stages)
mask_has_correct_start_0 = summary['has_correct_start'] == 0
summary.loc[mask_too_long_stages  | mask_has_correct_start_0, 'has_correct_length_stages'] = 0


# In fact only 3 uncorrected that are not matching YASA
summary['stage_match_without_shift'] = 1
mask_unmatch_yasa = summary.set_index(['sub_id', 'session']).index.isin(yasa_unmatching_stages)
mask_has_correct_start_0 = summary['has_correct_start'] == 0
summary.loc[mask_unmatch_yasa  | mask_has_correct_start_0, 'stage_match_without_shift'] = 0


summary = summary[['sub_id', 'session','has_edf', 'diagnostic', 'has_ecg', 'long_recording', 'has_sleep_stages', 'has_valid_sleep_stages', 'has_correct_start', 'has_correct_length_stages', 'stage_match_without_shift']]
summary = summary.sort_values(
    by=['has_edf', 'diagnostic', 'has_ecg', 'long_recording', 'has_sleep_stages', 'has_valid_sleep_stages', 'has_correct_start', 'has_correct_length_stages', 'stage_match_without_shift', 'sub_id', 'session'],
    ascending=[False, False, False, False, False, False, False, False, False, True, True]
)

summary = summary.fillna(0.0).reset_index(drop=True)
summary
```
