### Check the h5 file obtained

```python
import os
import h5py
import pandas as pd

def check_ecg_in_h5_folder(folder_path):
    results = []

    df_check = pd.read_csv("/Users/alicealbrecht/wynton_home/PSG_Pipeline/notebooks/h5_mgb_first_check.csv")
    df_check = df_check[~(df_check == False).any(axis=1)]
    already_check = set(df_check['file'])

    # Collect all .h5 files
    h5_files = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if (file not in already_check) and file.endswith(".h5"):
                h5_files.append(os.path.join(root, file))
            else:
                print("ALREADY CHECK: ", file)
   
    print(f"Found {len(h5_files)} .h5 files in {folder_path}\n")

    # Check each file
    for path in h5_files:
        status = {
            "file": os.path.basename(path),
            "path": path,
            "has_signals": False,
            "has_ecg": False,
            "ecg_nonempty": False,
            "error": None
        }

        try:
            with h5py.File(path, "r") as f:
                # 1. Check 'signals' group
                if "signals" in f:
                    status["has_signals"] = True

                    # 2. Check 'signals/ECG'
                    if "ECG" in f["signals"]:
                        status["has_ecg"] = True
                        ecg_group = f["signals"]["ECG"]

                        # 3. Check if not empty
                        if len(ecg_group.keys()) > 0:
                            status["ecg_nonempty"] = True
                else:
                    status["error"] = "Missing 'signals' group"

        except Exception as e:
            status["error"] = str(e)

        results.append(status)

    # Convert to DataFrame for easy inspection
    df = pd.DataFrame(results)

    # Print summary
    print("\n=== Summary ===")
    total = len(df)
    good = df["ecg_nonempty"].sum()
    print(f" {good}/{total} files have non-empty signals/ECG")
    print(f"  {total - good} files missing or empty ECG data\n")

    return df

# Example usage:
folder = "/Users/alicealbrecht/wynton_data/PSG_Pipeline_Outputs/h5_data/hsp_mgb"
df_check = check_ecg_in_h5_folder(folder)
```

```python
import os
import h5py

def clean_h5_folder(folder_path, recursive=False, dry_run=True):
    """
    Delete .h5 files missing signals or ECG, or with empty ECG group.

    Parameters
    ----------
    folder_path : str
        Folder containing .h5 files.
    recursive : bool
        If True, search subfolders.
    dry_run : bool
        If True, only print files that would be deleted.
        Set to False to actually delete them.
    """
    h5_files = []

    # Collect all .h5 files
    for root, dirs, files in os.walk(folder_path):
        for file in files[25000:]:
            if file.endswith(".h5"):
                h5_files.append(os.path.join(root, file))
        if not recursive:
            break

    print(f"Found {len(h5_files)} .h5 files in {folder_path}\n")

    # Check and delete files
    deleted_count = 0
    for path in h5_files:
        delete_file = False
        try:
            with h5py.File(path, "r") as f:
                if "signals" not in f:
                    delete_file = True
                elif "ECG" not in f["signals"]:
                    delete_file = True
                elif len(f["signals"]["ECG"].keys()) == 0:
                    delete_file = True
        except Exception as e:
            print(f"[ERROR] Cannot open {path}: {e}")
            delete_file = True

        if delete_file:
            if dry_run:
                print(f"[DRY RUN] Would delete: {path}")
            else:
                os.remove(path)
                print(f"Deleted: {path}")
            deleted_count += 1

    print(f"\nTotal files deleted (or would be deleted in dry run): {deleted_count}")

# Example usage:
folder = "/Users/alicealbrecht/wynton_data/PSG_Pipeline_Outputs/h5_data/hsp_mgb"
# dry_run=True just prints what would be deleted
clean_h5_folder(folder, recursive=False, dry_run=False)
```
