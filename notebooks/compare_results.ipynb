```python
from pathlib import Path
import pandas as pd

my_results_path = Path("/Users/alicealbrecht/Desktop/PSG_Pipeline_Outputs/extracted_features/mros_ses-1/run_20250919")
box_results = Path("/Users/alicealbrecht/Desktop/UCSF/pipeline_project/additional_data/box_results/")

stage_types = ["WN", "REM", "N2N3"]
all_df = []
for stage in stage_types:
    # Find file in my_results containing stage_type
    my_file = next(my_results_path.glob(f"*{stage}*.csv"), None)
    box_file = next(box_results.glob(f"*{stage}*.csv"), None)
    
    if my_file is None or box_file is None:
        print(f"[WARNING] No file found for stage {stage}, skipping...")
        continue
    
    # Read CSVs
    df_my = pd.read_csv(my_file)
    df_box = pd.read_csv(box_file)
    
    # Truncate box file to match number of rows of my results
    n_rows = len(df_my)
    df_box_trunc = df_box.iloc[:n_rows].copy()

    # Match cols
    new_cols = []
    for col in df_my.columns:
        if col != 'sub_id' and col.endswith(f"@{stage}"):
            new_cols.append(col.replace(f"@{stage}", ""))
        else:
            new_cols.append(col)
    df_my.columns = new_cols
    if 'sub_id' in df_my.columns:
        df_my = df_my.rename(columns={'sub_id': 'SID'})
    
    # Compute difference on common columns
    common_cols = df_my.columns.intersection(df_box_trunc.columns)
    cols_to_convert = [c for c in common_cols if c != "SID"]
    df_my[cols_to_convert] = df_my[cols_to_convert].astype(float)
    df_box_trunc[cols_to_convert] = df_box_trunc[cols_to_convert].astype(float)

    df_diff = df_my[cols_to_convert] - df_box_trunc[cols_to_convert]
    
    # Keep SID
    if 'SID' in df_my.columns:
        df_diff['SID'] = df_my['SID']
    
    all_df.append(df_diff)
```
